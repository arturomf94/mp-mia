```python
from __future__ import division
import networkx as nx
import random
import matplotlib.pyplot as plt
import math
```

```python
def edge_list(G):
	edges = [e for e in G.edges()]
	return edges
```

```python
def min_cut(G):
	while G.number_of_nodes() > 2: 
		edge = random.choice(edge_list(G))
		G = nx.contracted_edge(G,edge, self_loops = False)
	return G
```

```python
G1 = nx.MultiGraph()
G1.add_nodes_from(list(range(1,11)))
G1.add_edges_from([(1,2),(1,3),(1,4),(1,5),(2,3),(2,4),(2,5),(3,4),(3,5),(4,5),(6,7),(6,8),(6,9),(6,10),(7,8),(7,9),(7,10),(8,9),(8,10),(9,10),(5,6),(3,7),(4,8)])
```

```python
nx.draw(G1, with_labels = True)
plt.show()
```

```python
G_cut = min_cut(G1)
```

```python
nx.draw(G_cut, with_labels = True)
plt.show()
```

```python
def iterated_min_cut(G):
    min_edges = []
    n = G.number_of_nodes()
    iterations = int(math.ceil(n*(n-1)*math.log(n)))
    for i in range(0,iterations):
        G_cut = min_cut(G)
        min_edges.append(G_cut.number_of_edges())
    return min(min_edges)
```

```python
print(iterated_min_cut(G1))
```

```python
def bound(n):
    return (1-2/(n*(n-1)))**(n*(n-1)*math.log(n))
```
